**Import necessery libraries**

```python
import pandas as pd
```

**Load the dataset**

```python
df = pd.read_csv("/content/fraud test.csv")
```

**Know about the loaded dataset**

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

**Data Preprocessing**

```python
df.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'], format='%d/%m/%Y %H:%M', errors='coerce')
df['Trans_Date'] = df['trans_date_trans_time'].dt.date
df['Trans_Time'] = df['trans_date_trans_time'].dt.time
```

```python
df.drop(['trans_date_trans_time'], inplace=True, axis=1)
```

```python
df['dob'] = pd.to_datetime(df['dob'])
```

```python
df['is_fraud'] = df['is_fraud'].astype(int)
df['zip'] = df['zip'].astype(int)
df['city_pop'] = df['city_pop'].astype(int)
df['zip'] = df['zip'].astype(int)
```

```python
df['location'] = df['city'] + ', ' + df['state']
df.drop(['city','state'], axis = 1, inplace = True)
```

```python
df['gender'] = df['gender'].replace({'M': 0, 'F': 1})
df['gender'] = df['gender'].astype(int)
```

**Data Transformation**

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Trans_Date'] = label_encoder.fit_transform(df['Trans_Date'])
df['Trans_Time'] = label_encoder.fit_transform(df['Trans_Time'])
df['trans_num'] = label_encoder.fit_transform(df['trans_num'])
df['job'] = label_encoder.fit_transform(df['job'])
df['dob'] = label_encoder.fit_transform(df['dob'])
df['cc_num'] = label_encoder.fit_transform(df['cc_num'])
df['merchant'] = label_encoder.fit_transform(df['merchant'])
df['street'] = label_encoder.fit_transform(df['street'])
df['category'] = label_encoder.fit_transform(df['category'])
df['location'] = label_encoder.fit_transform(df['location'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, cmap='ocean', fmt=".2f", annot_kws={"size": 10})

plt.title('Correlation Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
x = df.drop(columns=['is_fraud', 'dob', 'first', 'last', 'category', 'trans_num'], axis=1)
y= df['is_fraud']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

**Model Building**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=16)
```

```python
from imblearn.over_sampling import SMOTE
print("Before Over Sampling, count of the label '1': {}".format(sum(y_train == 1)))
print("Before Over Sampling, count of the label '0': {} \n".format(sum(y_train == 0)))

sm1 = SMOTE(random_state=2)
x_train_res, y_train_res = sm1.fit_resample(x_train, y_train)

print('After Over Sampling, the shape of the train_X: {}'.format(x_train_res.shape))
print('After Over Sampling, the shape of the train_y: {} \n'.format(y_train_res.shape))

print("After Over Sampling, count of the label '1': {}".format(sum(y_train_res == 1)))
print("After Over Sampling, count of the label '0': {}".format(sum(y_train_res == 0)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LogisticRegression()
model.fit(x_train_res, y_train_res)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
train_predictions = model.predict(x_train_res)
train_accuracy = accuracy_score(y_train_res, train_predictions)

test_accuracy = accuracy_score(y_test, predict)

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```

**Ada Boost**

```python
from sklearn.ensemble import AdaBoostClassifier

Ada_boost_model=AdaBoostClassifier()
Ada_boost_model.fit(x_train_res, y_train_res )

y_pred_Ada= Ada_boost_model.predict(x_test)
y_pred_train=Ada_boost_model.predict(x_train_res)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,classification_report
print("Training_Accuracy :",(accuracy_score(y_train_res,y_pred_train)))
print("Confusion_Matrix : \n",confusion_matrix(y_train_res,y_pred_train))
print("Classification_Report : \n",classification_report(y_train_res,y_pred_train))
```

```python
print("Testing_Accuracy :",(accuracy_score(y_test,y_pred_Ada)))
print("Confusion_Matrix : \n",confusion_matrix(y_test,y_pred_Ada))
print("Classification_Report : \n",classification_report(y_test,y_pred_Ada))
```

**Xg Boost**

```python
from xgboost import XGBClassifier
xgb_model=XGBClassifier()
xgb_model.fit(x_train_res, y_train_res )

y_predict_xgb= xgb_model.predict(x_test)
xgb_pred_train=xgb_model.predict(x_train_res)
```

```python
print("Training_Accuracy :",(accuracy_score(y_train_res,xgb_pred_train)))
print("Confusion_Matrix : \n",confusion_matrix(y_train_res,xgb_pred_train))
print("Classification_Report : \n",classification_report(y_train_res,xgb_pred_train))
```

```python
print("Testing_Accuracy :",(accuracy_score(y_test,y_predict_xgb)))
print("Confusion_Matrix : \n",confusion_matrix(y_test,y_predict_xgb))
print("Classification_Report : \n",classification_report(y_test,y_predict_xgb))
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model architecture
model = Sequential([
    Dense(64, activation='relu', input_shape=(16,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


# Train the model
model_train = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
train_accuracy = model_train.history['accuracy'][-1]

print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
print(f'Train Accuracy: {train_accuracy}')

```
